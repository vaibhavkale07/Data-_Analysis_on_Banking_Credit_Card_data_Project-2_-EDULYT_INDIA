### Create a profile of customers in terms of timing of their order. 

```python
morning_segment = (datetime.strptime('06:00:00', '%H:%M:%S').time(), datetime.strptime('11:59:59', '%H:%M:%S').time())
evening_segment = (datetime.strptime('12:00:00', '%H:%M:%S').time(), datetime.strptime('17:59:59', '%H:%M:%S').time())
night_segment = (datetime.strptime('18:00:00', '%H:%M:%S').time(), datetime.strptime('05:59:59', '%H:%M:%S').time())
```

```python
new_df['Time'] = pd.to_datetime(new_df['Time'], format='%H:%M:%S')
```

```python
new_df['Time_Segment'] = new_df['Time'].apply(lambda x: 'Morning' if morning_segment[0] <= x.time() <= morning_segment[1]
                                      else ('Evening' if evening_segment[0] <= x.time() <= evening_segment[1]
                                            else 'Night'))
```
